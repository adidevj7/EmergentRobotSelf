```python
# JUPYTER CELL — Batch plot + DISPLAY ALL graphs inline (depth ≤ 2)
# Saves PNG next to each rollout_log.csv AND shows them in the notebook.

from __future__ import annotations
from pathlib import Path
from typing import List, Optional, Tuple
import re

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from matplotlib.legend_handler import HandlerBase
import seaborn as sns
from IPython.display import display, Image

# ──────────────────────────────────────────────────────────────────────────────
# INPUTS
# ──────────────────────────────────────────────────────────────────────────────
ROOT_DIR  = Path("/Users/adi/Desktop/EmergentRobotSelf/Checkpoints_States_selectedGraphs/WalkSpinJump_relu")
MAX_DEPTH = 2
N_CKPTS   = 12
FIGSIZE   = (13, 5)

PHASE_LABELS = ["Walk", "Wiggle", "Bob"]          # repeats
CKPT_PAT     = re.compile(r"\bckpt\b", re.IGNORECASE)

# If True: show inline (matplotlib figs). If False: show saved PNGs inline.
SHOW_FIGS_INLINE = True

# ──────────────────────────────────────────────────────────────────────────────
# Style
# ──────────────────────────────────────────────────────────────────────────────
sns.set_theme(style="whitegrid", context="talk")
sns.set_context("talk", font_scale=0.95)

# ──────────────────────────────────────────────────────────────────────────────
# Helpers
# ──────────────────────────────────────────────────────────────────────────────
def _stitch_monotone(series: pd.Series) -> pd.Series:
    x = pd.to_numeric(series, errors="coerce").to_numpy(dtype=float)
    if x.size == 0:
        return pd.Series(x, index=series.index)
    offsets = np.zeros_like(x)
    running_offset = 0.0
    for i in range(1, x.size):
        if not np.isfinite(x[i]) or not np.isfinite(x[i - 1]):
            offsets[i] = running_offset
            continue
        if x[i] < x[i - 1]:
            running_offset += max(0.0, x[i - 1] - x[i])
        offsets[i] = running_offset
    return pd.Series(x + offsets, index=series.index)

def load_rollout_df(path: str | Path, stitch: bool) -> pd.DataFrame:
    df = pd.read_csv(Path(path))
    df.columns = [c.strip() for c in df.columns]

    required = ["time_s", "episodes_done", "total_steps", "avg_return_window", "avg_len_window"]
    missing = [c for c in required if c not in df.columns]
    if missing:
        raise ValueError(f"Missing required columns: {missing}")

    for c in required:
        df[c] = pd.to_numeric(df[c], errors="coerce")

    df = df.dropna(subset=["total_steps", "avg_return_window"]).reset_index(drop=True)

    if stitch:
        df["time_s"]        = _stitch_monotone(df["time_s"])
        df["episodes_done"] = _stitch_monotone(df["episodes_done"])
        df["total_steps"]   = _stitch_monotone(df["total_steps"])

    return df

def parse_first_n_ckpt_row_indices(csv_path: str | Path, n_ckpts: int) -> List[int]:
    """
    Return dataframe-row indices where 'ckpt' markers occur (first n_ckpts).
    Row-index parsing avoids duplicate raw step ambiguity.
    """
    text = Path(csv_path).read_text(encoding="utf-8", errors="ignore")
    ckpt_row_idxs: List[int] = []

    data_row_idx = -1
    for line in text.splitlines():
        s = line.strip()
        if not s:
            continue

        parts = [x.strip() for x in s.split(",")]
        if parts and parts[0].lower() == "time_s":
            continue

        # Count data rows
        is_data_row = False
        if len(parts) >= 3:
            try:
                float(parts[0])      # time_s
                float(parts[2])      # total_steps
                is_data_row = True
            except Exception:
                is_data_row = False
        if is_data_row:
            data_row_idx += 1

        # Marker line
        if CKPT_PAT.search(s):
            if data_row_idx >= 0:
                ckpt_row_idxs.append(data_row_idx)
                if len(ckpt_row_idxs) >= n_ckpts:
                    break

    return ckpt_row_idxs

def fmt_steps(x: float) -> str:
    x = float(x)
    ax = abs(x)
    if ax >= 1_000_000_000:
        return f"{x/1_000_000_000:.1f}B"
    if ax >= 1_000_000:
        return f"{x/1_000_000:.1f}M"
    if ax >= 1_000:
        return f"{x/1_000:.1f}k"
    return f"{int(x)}"

# Legend handler to draw a VERTICAL line key (not horizontal)
class VerticalLineHandler(HandlerBase):
    def create_artists(self, legend, orig_handle, xdescent, ydescent, width, height, fontsize, trans):
        x = xdescent + width * 0.5
        y0 = ydescent + height * 0.15
        y1 = ydescent + height * 0.85
        line = Line2D([x, x], [y0, y1],
                      color=orig_handle.get_color(),
                      linestyle=orig_handle.get_linestyle(),
                      linewidth=orig_handle.get_linewidth(),
                      alpha=orig_handle.get_alpha())
        line.set_transform(trans)
        return [line]

def find_rollout_logs(root: Path, max_depth: int) -> List[Path]:
    root = root.resolve()
    hits: List[Path] = []
    for p in root.rglob("rollout_log.csv"):
        try:
            rel = p.parent.relative_to(root)
        except Exception:
            continue
        if len(rel.parts) <= max_depth:
            hits.append(p)
    return sorted(hits)

def make_plot_for_csv(csv_path: Path, n_ckpts: int) -> Tuple[plt.Figure, Path]:
    df = load_rollout_df(csv_path, stitch=True)
    ckpt_row_idxs = parse_first_n_ckpt_row_indices(csv_path, n_ckpts)

    # Truncate to Nth checkpoint (inclusive)
    if ckpt_row_idxs:
        cut_idx = min(int(ckpt_row_idxs[-1]), len(df) - 1)
        df_plot = df.iloc[: cut_idx + 1].copy()
    else:
        df_plot = df.copy()

    # checkpoint x positions (stitched) + dedupe consecutive duplicates
    ckpt_x_all = [float(df["total_steps"].iloc[i]) for i in ckpt_row_idxs if 0 <= i < len(df)]
    ckpt_x: List[float] = []
    for x in ckpt_x_all:
        if not ckpt_x or abs(x - ckpt_x[-1]) > 1e-6:
            ckpt_x.append(x)

    # phase bounds: start + ckpts + end
    x0 = float(df_plot["total_steps"].iloc[0]) if len(df_plot) else 0.0
    x1 = float(df_plot["total_steps"].iloc[-1]) if len(df_plot) else 1.0
    bounds = [x0] + ckpt_x + [x1]

    # ── plot ─────────────────────────────────────────────────────────────────
    fig, ax = plt.subplots(figsize=FIGSIZE)

    ret_handle, = ax.plot(
        df_plot["total_steps"],
        df_plot["avg_return_window"],
        linewidth=2.6,
        alpha=0.95,
        color="tab:orange",
        label="Episode Return",
    )

    for x in ckpt_x:
        ax.axvline(x, color="black", linestyle=(0, (6, 6)), linewidth=1.6, alpha=0.9)

    ax.set_xlabel("Total Timesteps")
    ax.set_ylabel("Episode Return")
    ax.set_title("Return vs Steps — Example Rollout (First 2 Cycles)", fontweight="bold")

    # bottom labels (x alignment unchanged)
    ymin, ymax = ax.get_ylim()
    y_text = ymin + 0.10 * (ymax - ymin)

    for i in range(len(bounds) - 1):
        left, right = bounds[i], bounds[i + 1]
        if not (np.isfinite(left) and np.isfinite(right)) or right <= left:
            continue
        x_mid = 0.5 * (left + right)
        label = PHASE_LABELS[i % len(PHASE_LABELS)]
        ax.text(
            x_mid, y_text, label,
            ha="center", va="center",
            fontsize=12, fontweight="bold",
            color="black",
            bbox=dict(boxstyle="round,pad=0.25", facecolor="white", edgecolor="none", alpha=0.7),
            clip_on=True,
            zorder=10,
        )

    xt = ax.get_xticks()
    ax.set_xticklabels([fmt_steps(t) for t in xt])

    ckpt_proxy = Line2D([0], [0], color="black", linestyle=(0, (6, 6)), linewidth=1.8, alpha=0.9)
    ax.legend(
        handles=[ret_handle, ckpt_proxy],
        labels=["Episode Return", "Checkpoint"],
        loc="upper right",
        frameon=True,
        handler_map={ckpt_proxy: VerticalLineHandler()},
    )

    sns.despine(ax=ax, top=True, right=True)
    fig.tight_layout()

    out_png = csv_path.parent / "rollout_return_vs_steps_first2cycles.png"
    fig.savefig(out_png, dpi=200, bbox_inches="tight")
    return fig, out_png

# ──────────────────────────────────────────────────────────────────────────────
# Batch run + SHOW ALL
# ──────────────────────────────────────────────────────────────────────────────
logs = find_rollout_logs(ROOT_DIR, MAX_DEPTH)
print(f"[found] {len(logs)} rollout_log.csv files (depth ≤ {MAX_DEPTH}) under:\n  {ROOT_DIR}")

figs: List[plt.Figure] = []
pngs: List[Path] = []

for csv_path in logs:
    try:
        fig, out_png = make_plot_for_csv(csv_path, N_CKPTS)
        figs.append(fig)
        pngs.append(out_png)

        # Print a short header so you know which run you're looking at
        rel = csv_path.parent.relative_to(ROOT_DIR)
        print(f"[ok] {rel}  ->  {out_png.name}")

        if SHOW_FIGS_INLINE:
            plt.show(fig)
            plt.close(fig)  # keeps notebook memory sane
        else:
            plt.close(fig)
            display(Image(filename=str(out_png)))

    except Exception as e:
        rel = csv_path.parent.relative_to(ROOT_DIR) if csv_path.exists() else csv_path
        print(f"[fail] {rel}: {type(e).__name__}: {e}")

print(f"[done] displayed {len(pngs)} graphs, saved PNGs next to each rollout_log.csv")
```

```python
# JUPYTER CELL — Return vs Steps ONLY, truncated to first N checkpoints,
# add mid-phase labels (Walk / Wiggle / Bob repeating), and a TRUE vertical-line legend key.

from __future__ import annotations
from pathlib import Path
from typing import List, Optional, Tuple
import re

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from matplotlib.legend_handler import HandlerBase
import seaborn as sns

# ──────────────────────────────────────────────────────────────────────────────
# INPUTS
# ──────────────────────────────────────────────────────────────────────────────
LOG_PATH = "/Users/adi/Desktop/EmergentRobotSelf/Checkpoints_States_selectedGraphs/WalkSpinJump_relu/WSJ_relu_5_2026-01-30_23-28-40/rollout_log.csv"
N_CKPTS  = 12                      # we will truncate to the Nth ckpt marker
FIGSIZE  = (13, 5)

PHASE_LABELS = ["Walk", "Wiggle", "Bob"]  # repeats across phases in this fixed order

CKPT_PAT = re.compile(r"\bckpt\b", re.IGNORECASE)

# ──────────────────────────────────────────────────────────────────────────────
# Style
# ──────────────────────────────────────────────────────────────────────────────
sns.set_theme(style="whitegrid", context="talk")
sns.set_context("talk", font_scale=0.95)

# ──────────────────────────────────────────────────────────────────────────────
# Helpers
# ──────────────────────────────────────────────────────────────────────────────
def _stitch_monotone(series: pd.Series) -> pd.Series:
    x = pd.to_numeric(series, errors="coerce").to_numpy(dtype=float)
    if x.size == 0:
        return pd.Series(x, index=series.index)
    offsets = np.zeros_like(x)
    running_offset = 0.0
    for i in range(1, x.size):
        if not np.isfinite(x[i]) or not np.isfinite(x[i - 1]):
            offsets[i] = running_offset
            continue
        if x[i] < x[i - 1]:
            running_offset += max(0.0, x[i - 1] - x[i])
        offsets[i] = running_offset
    return pd.Series(x + offsets, index=series.index)

def load_rollout_df(path: str | Path, stitch: bool) -> pd.DataFrame:
    df = pd.read_csv(Path(path))
    df.columns = [c.strip() for c in df.columns]

    required = ["time_s", "episodes_done", "total_steps", "avg_return_window", "avg_len_window"]
    missing = [c for c in required if c not in df.columns]
    if missing:
        raise ValueError(f"Missing required columns: {missing}")

    for c in required:
        df[c] = pd.to_numeric(df[c], errors="coerce")

    df = df.dropna(subset=["total_steps", "avg_return_window"]).reset_index(drop=True)

    if stitch:
        df["time_s"]        = _stitch_monotone(df["time_s"])
        df["episodes_done"] = _stitch_monotone(df["episodes_done"])
        df["total_steps"]   = _stitch_monotone(df["total_steps"])

    return df

def parse_first_n_ckpt_row_indices(csv_path: str | Path, n_ckpts: int) -> Tuple[List[int], List[int]]:
    """
    Parse RAW CSV text and return dataframe-row indices where 'ckpt' markers occur.
    Uses row indices so duplicate raw step values don't break placement.
    """
    text = Path(csv_path).read_text(encoding="utf-8", errors="ignore")

    ckpt_row_idxs: List[int] = []
    ckpt_raw_steps: List[int] = []

    data_row_idx = -1
    last_steps: Optional[int] = None

    for line in text.splitlines():
        s = line.strip()
        if not s:
            continue

        parts = [x.strip() for x in s.split(",")]
        if parts and parts[0].lower() == "time_s":
            continue

        is_data_row = False
        if len(parts) >= 3:
            try:
                float(parts[0])
                last_steps = int(float(parts[2]))
                is_data_row = True
            except Exception:
                is_data_row = False

        if is_data_row:
            data_row_idx += 1

        if CKPT_PAT.search(s):
            if data_row_idx >= 0:
                ckpt_row_idxs.append(data_row_idx)
                ckpt_raw_steps.append(last_steps if last_steps is not None else -1)
                if len(ckpt_row_idxs) >= n_ckpts:
                    break

    return ckpt_row_idxs, ckpt_raw_steps

def fmt_steps(x: float) -> str:
    x = float(x)
    ax = abs(x)
    if ax >= 1_000_000_000:
        return f"{x/1_000_000_000:.1f}B"
    if ax >= 1_000_000:
        return f"{x/1_000_000:.1f}M"
    if ax >= 1_000:
        return f"{x/1_000:.1f}k"
    return f"{int(x)}"

# ──────────────────────────────────────────────────────────────────────────────
# Legend handler to draw a VERTICAL line key (not horizontal)
# ──────────────────────────────────────────────────────────────────────────────
class VerticalLineHandler(HandlerBase):
    def create_artists(self, legend, orig_handle, xdescent, ydescent, width, height, fontsize, trans):
        x = xdescent + width * 0.5
        y0 = ydescent + height * 0.15
        y1 = ydescent + height * 0.85
        line = Line2D([x, x], [y0, y1],
                      color=orig_handle.get_color(),
                      linestyle=orig_handle.get_linestyle(),
                      linewidth=orig_handle.get_linewidth(),
                      alpha=orig_handle.get_alpha())
        line.set_transform(trans)
        return [line]

# ──────────────────────────────────────────────────────────────────────────────
# Run
# ──────────────────────────────────────────────────────────────────────────────
logp = Path(LOG_PATH).expanduser()
if not logp.exists():
    raise FileNotFoundError(f"LOG_PATH does not exist: {logp}")

df = load_rollout_df(logp, stitch=True)
ckpt_row_idxs, ckpt_raw_steps = parse_first_n_ckpt_row_indices(logp, N_CKPTS)

# Truncate to Nth checkpoint (inclusive)
if ckpt_row_idxs:
    cut_idx = min(int(ckpt_row_idxs[-1]), len(df) - 1)
    df_plot = df.iloc[: cut_idx + 1].copy()
else:
    df_plot = df.copy()

# Checkpoint x positions from row indices, then dedupe consecutive duplicates
ckpt_x_all = [float(df["total_steps"].iloc[i]) for i in ckpt_row_idxs if 0 <= i < len(df)]
ckpt_x: List[float] = []
for x in ckpt_x_all:
    if not ckpt_x or abs(x - ckpt_x[-1]) > 1e-6:
        ckpt_x.append(x)

# Phase boundaries: start at left edge, then each ckpt line, then right edge
x0 = float(df_plot["total_steps"].iloc[0]) if len(df_plot) else 0.0
x_end = float(df_plot["total_steps"].iloc[-1]) if len(df_plot) else 1.0
bounds = [x0] + ckpt_x + [x_end]

# ──────────────────────────────────────────────────────────────────────────────
# Plot
# ──────────────────────────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=FIGSIZE)

# Return line in orange
ret_handle, = ax.plot(
    df_plot["total_steps"],
    df_plot["avg_return_window"],
    linewidth=2.6,
    alpha=0.95,
    color="tab:orange",
    label="Episode Return",
)

# Checkpoint lines
for x in ckpt_x:
    ax.axvline(x, color="black", linestyle=(0, (6, 6)), linewidth=1.6, alpha=0.9)

# Mid-phase labels: Walk, Wiggle, Bob repeating (no parsing; just sequential)
ymin, ymax = ax.get_ylim()
y_text = ymin + 0.92 * (ymax - ymin)  # near top inside axes
for i in range(len(bounds) - 1):
    left, right = bounds[i], bounds[i + 1]
    if not (np.isfinite(left) and np.isfinite(right)) or right <= left:
        continue
    x_mid = 0.5 * (left + right)
    label = PHASE_LABELS[i % len(PHASE_LABELS)]
    ax.text(
        x_mid, y_text, label,
        ha="center", va="center",
        fontsize=12, fontweight="bold",
        color="black",
        bbox=dict(boxstyle="round,pad=0.25", facecolor="white", edgecolor="none", alpha=0.7),
        clip_on=True,
        zorder=10,
    )

ax.set_xlabel("Total Timesteps")
ax.set_ylabel("Episode Return")
ax.set_title("Return vs Steps — Example Rollout (First 2 Cycles)", fontweight="bold")

# Pretty x tick labels
xt = ax.get_xticks()
ax.set_xticklabels([fmt_steps(t) for t in xt])

# Legend: real vertical-line key via custom handler
ckpt_proxy = Line2D([0], [0], color="black", linestyle=(0, (6, 6)), linewidth=1.8, alpha=0.9)
ax.legend(
    handles=[ret_handle, ckpt_proxy],
    labels=["Episode Return", "Checkpoint"],
    loc="upper right",
    frameon=True,
    handler_map={ckpt_proxy: VerticalLineHandler()},
)

sns.despine(ax=ax, top=True, right=True)
fig.tight_layout()
plt.show()

print(f"[info] LOG_PATH: {logp}")
print(f"[info] ckpt row indices: {ckpt_row_idxs}")
print(f"[info] ckpt raw steps: {ckpt_raw_steps}")
print(f"[info] ckpt plotted x (deduped): {np.round(ckpt_x, 2).tolist()}")
print(f"[info] plotted rows: 0..{len(df_plot)-1} (of {len(df)})")
```
